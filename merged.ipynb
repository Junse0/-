sampling rate을 통해 sin 곡선을 시간과 연결시킬 수 있음
(radian 정보를 시간 정보로 바꿀 수 있음)

문제: sampling rate이 10Hz면, 한 sample이 의미하는 초는 0.1초

sampling rate이 10000Hz면, 한 sample이 담당하는 초는 10000분의 1초

sampling rate은 resolution(해상도)에 영향을 줌

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd                   #소리를 play할 때 사용
```

### Phasor

1단계: sample index 만들기

2단계: sample index를 time으로 변환

3단계: time을 angle로 변환

4단계: angle을 sin 곡선에 넣음

```python
# parameter setting
amp = 1          # range [0.0, 1.0]     # 시간이 개입된 sin곡선의 amplitude는 -1에서 1까지 감(0에서 1까지가 아님)
sr = 10000       # sampling rate, Hz
dur = 0.5        # in seconds           # 0.5초의 샘플은 몇 번째 샘플? 5000번째 샘플
freq = 100.0     # sine frequency, Hz   # 0.5초 안에는 50번의 사이클이 있음
phase = np.pi/2  # phase, radian        # 90도 부터 시작하겠다.(cos과 같음)
```

만약, 

amp=2라면 그 sin 곡선의 최솟값은 -2

```python
# generate samples
nSamp = sr * dur                        # 만들려고 하는 샘플 개수(5000개)
samples = np.arange(1, nSamp+1)         # 1에서 5000까지(마지막 포함 안 함) 샘플 인덱스(넘버)를 만듦
```

```python
# convert samples to time
t = samples/sr                          # 1~ 5000 / 시간 = 1/10000 ~ 5000/10000 (총 5000개의 시간 샘플-time ticks-을 만듦)
```

```python
# convert time to angle in radian
theta0 = t * 2*np.pi * freq             # time을 angle로 만듦(sin에는 각도를 넣어야 하므로) #theta0(0도 부터 시작한다는 의미로...)
```

ㄴ freq=1 일때,

[0초...1초] * 2π -> [0...2π]

freq=100 일때는, 

[0초...1초] * 2π -> [0...2π]*100

추가
freq가 1이라는 것은 2π가 한 번 있다
freq가 100이라는 것은 2π가 1초에 100번 있다

```python
# shift angle by phase
theta = theta0 + phase    # phase shift
```

```python
# generate signal by sin-phasor
s = amp * np.sin(theta)
```

```python
# plot the entire signal    # time function
plt.plot(t, s, ':')   #x축, y축
plt.xlabel('time (s)')
plt.ylabel('real')
```

```python
# x값에 theta가 들어간다면...
plt.plot(theta, s, ':')   # x축, y축
plt.xlabel('angle(radian)')
plt.ylabel('real')
# 이 그래프의 문제점 -> 현실적으로 몇 초인지 알 수 없음
# 모양은 똑같이 보임(time의 첫번째 값 1/10000 theta의 첫번째 값 1/10000*2π*freq)
```

```python
def phasor_sine(amp, sr, dur, freq, phase):     # 앞 과정의 function 만들기    # 입력: amp, sr, dur, freq, phase
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta) 
  return s, t, theta                            # 출력: s, t, theta (새로 계산된 값이기도 함)  
```

### Dividing signal by the window size and step

```python
def frame_signal(signal, srate, win_size, win_step):    # framing: amplitude를 측정하기 위해서 slice해서 size를 잼  
                                                        # win_size(초): 창의 크기  # win_step(초): 옆으로 얼마나 가는가   
                                                        # win_step이 작으면 resolution이 커짐(조밀)
    frame_size = int(win_size * srate)
    frame_step = int(win_step * srate)

    # Get number of frames
    num_frames = int(np.ceil(np.abs(len(signal) - frame_size) / frame_step))
    # Pad signal
    pad_len = num_frames * frame_step + frame_size
    pad = np.zeros((pad_len - len(signal)))
    signal_padded = np.append(signal, pad)
    # Get within-frame sample indices
    idx1 = np.tile(np.arange(0, frame_size), (num_frames, 1))
    # Get vectors of frame_step increments
    idx2 = np.tile(np.arange(0, num_frames * frame_step, frame_step),
                   (frame_size, 1)).T
    # Get total indices divided by each frame
    indices = idx1 + idx2
    # Get frames divided by each frame based on indices
    frames = signal_padded[indices.astype(np.int32, copy=False)]
#     if cutoff is not None:
#         cutoff_bin = round(cutoff * nfft / (srate))  # hz -> bin
#         frames = frames[:, :cutoff_bin]
    return frames

# 입력과 출력만 보면 됨
# 입력(signal, srate, win_size, win_step)
# 출력 (frames) 즉 frame이란 array를 뽑아줌
```

### Intensity by RMS(RMS: root mean square)

```python
amp = []                                              # 빈 list 생성
for frame in frames:                                  # frames는 ndarray with shape (38, 2205)의 행렬 # 이렇게 list(벡터) 대신 행렬을 넣으면 앞 부분(38에 해당)하는 것을 뽑아감 # 8 만 개 중 2200개를 자름
  amp.append(np.sqrt(np.mean(frame**2)))              # 2200개 자른 것(frame)의 모든 값을 제곱하고, 평균을 치고(np.mean), 루트를 씌움(np.sqrt) => intensity를 구할 수 있음
# ㄴ .append로 위에서 생성한 빈 list에 하나하나씩 채움(38번의 for loop)

plt.subplot(2,1,1)
plt.plot(t, s)
plt.xlim([0, t[-1]])                                  # x의 range   #t[-1] time의 끝
plt.subplot(2,1,2)
plt.plot(framesT, amp, '.')                           # framesT와 amp의 개수가 같아야 함(38개)
plt.xlim([0, t[-1]])
plt.show()
```

### Pitch by autocorrelation*(시험에 포함되지 않음)

```python
def autocorr(s):
    result = np.correlate(s, s, mode='full')
    maxID = np.argmax(result)
    return result[maxID:]/result[maxID]
```

```python
F0 = []
for frame in frames:    #intensity by RMS와 마찬가지로 frame을 이용함
  result = autocorr(frame)
  peakIdx = np.argmin(result) + np.argmax(result[np.argmin(result):])
  if result[peakIdx] >= 0.5:
    F0.append(1/(peakIdx/sr))
  else:
    F0.append(np.nan)

plt.subplot(2,1,1)
plt.plot(t, s)
plt.xlim([0, t[-1]])
plt.subplot(2,1,2)
plt.plot(framesT, F0, '.')
plt.xlim([0, t[-1]])
plt.show()
```

```python
sent = "I am Jhon from America and would like to go to Starbucks"   
words = nltk.word_tokenize(sent)
```

```python
'''
POS tag list:

CC	coordinating conjunction
CD	cardinal digit
DT	determiner
EX	existential there (like: "there is" ... think of it like "there exists")
FW	foreign word
IN	preposition/subordinating conjunction
JJ	adjective	'big'
JJR	adjective, comparative	'bigger'
JJS	adjective, superlative	'biggest'
LS	list marker	1)
MD	modal	could, will
NN	noun, singular 'desk'
NNS	noun plural	'desks'
NNP	proper noun, singular	'Harrison'
NNPS	proper noun, plural	'Americans'
PDT	predeterminer	'all the kids'
POS	possessive ending	parent's
PRP	personal pronoun	I, he, she
PRP$	possessive pronoun	my, his, hers
RB	adverb	very, silently,
RBR	adverb, comparative	better
RBS	adverb, superlative	best
RP	particle	give up
TO	to	go 'to' the store.
UH	interjection	errrrrrrrm
VB	verb, base form	take
VBD	verb, past tense	took
VBG	verb, gerund/present participle	taking
VBN	verb, past participle	taken
VBP	verb, sing. present, non-3d	take
VBZ	verb, 3rd person sing. present	takes
WDT	wh-determiner	which
WP	wh-pronoun	who, what
WP$	possessive wh-pronoun	whose
WRB	wh-abverb	where, when
'''
```

```python
nltk.download('averaged_perceptron_tagger')   #pos(품사) tagger를 다운
pos = nltk.pos_tag(words)   # 품사 정보가 tag된 상태
pos
```

```python
nltk.download('maxent_ne_chunker')
NE = nltk.ne_chunk(pos)   # ne: named entity(ORGANIZATION, PERSON, LOCATION, DATE, TIME, MONEY, and GPE를 tagging해줌)
# common Entity types: ORGANIZATION, PERSON, LOCATION, DATE, TIME, MONEY, and GPE (geo-political entity)
```
